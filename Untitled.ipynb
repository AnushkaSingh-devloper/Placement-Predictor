```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("placement.csv")
```

```python
df = df.iloc[:,1:]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
placed = df[df['placement'] == 1]  # separate data by class
not_placed = df[df['placement'] == 0]

plt.scatter(placed['cgpa'],placed['iq'] ,color = 'green', label = 'placed')
plt.scatter(not_placed['cgpa'],not_placed['iq'] ,color = 'red', label = 'not_placed')

plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.legend()
plt.show()
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() # class object
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# model training
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
import pickle
```

```python
pickle.dump(clf,open('placement_model.pkl','wb'))
```
